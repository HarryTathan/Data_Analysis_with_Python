```python
import pandas as pd
```

```python
df = pd.read_excel("flight_price.xlsx")
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.shape
```

```python
df.head()
```

```python
#Splitting Date
df["date"] = df['Date_of_Journey'].str.split('/').str[0]
df["month"] = df['Date_of_Journey'].str.split('/').str[1]
df["year"] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
# Converting date to int.
df["date"] = df['date'].astype(int)
df["month"] = df['month'].astype(int)
df["year"] = df['year'].astype(int)
```

```python
df.drop("Date_of_Journey",axis=1,inplace=True)
```

```python
df["arival_time"]=df['Arrival_Time'].str.split(' ').str[0]
```

```python
df.head()
```

```python
df.drop("Arrival_Time", axis = 1, inplace = True)
```

```python
df.head()
```

```python
df.info()
```

```python
df["dep_hr"] = df["Dep_Time"].str.split(":").str[0]
df["dep_min"] = df["Dep_Time"].str.split(":").str[1]
```

```python
df["arr_hr"]=df["arival_time"].str.split(":").str[0]
df["arr_min"]=df["arival_time"].str.split(":").str[1]
```

```python
df.drop("Dep_Time",axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Total_Stops'].unique()
```

```python
# Using Map function to update Total stops column

df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4})
```

```python
df.drop("Route", axis = 1, inplace = True)
```

```python
df.head()
```

```python
df["duration_hr"]=df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df["duration_min"]=df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
df.head()
```

```python
df['duration_min'].isnull().sum()
```

```python
df["duration_min"].fillna(0,inplace=True)
```
