```python
import pandas as pd
```

```python
df = pd.read_csv("googleplaystore.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# ~ used for not
df[~df['Reviews'].str.isnumeric()]
```

```python
df.drop(10472, inplace=True) # non - numeric
```

```python
df.shape
```

```python
df['Reviews']=df['Reviews'].astype(int)
```

```python
df['Size'].unique()
```

```python
# Converting to same units
df['Size'] = df['Size'].str.replace('M','000k')
```

```python
df["Size"].unique()
```

```python
df['Size'] = df['Size'].str.replace('k','')
```

```python
df["Size"].unique()
```

```python
df['Size']=df['Size'].str.replace('Varies with device','NaN')
```

```python
df['Size']=df['Size'].astype(float)
```

```python
df.info()
```

```python
df.head()
```

```python
chars_to_remove = ["+",",","$"]
cols_to_remove = ["Installs","Price"]
for col in cols_to_remove:
    for char in chars_to_remove:
        df[col] = df[col].str.replace(char,"")
```

```python
df['Installs']=df['Installs'].astype(int)
```

```python
df['Price']=df['Price'].astype(float)
```

```python
df.head()
```

```python
df['Last Updated']=pd.to_datetime(df['Last Updated'])
df["Day"]=df['Last Updated'].dt.day
df["Month"]=df['Last Updated'].dt.month
df["Year"]=df['Last Updated'].dt.year
```

```python
df.head()
```

```python
df.drop("Last Updated",axis=1,inplace=True)
```

```python
df.head()
```

```python
df.duplicated("App").sum()
```

```python
# Drop Duplicates
df.drop_duplicates("App",inplace=True)
```

```python
df.duplicated("App").sum()
```

```python
df.isnull().sum()
```

```python
df['Size'].median()
```

```python
df['Size'].mean()
```

```python
df['Size'] = df['Size'].fillna(df['Size'].median())
```

```python
df['Rating'].unique()
```

```python
df['Rating']=df['Rating'].fillna(df['Rating'].median())
```

```python
numerical_features = [feature for feature in df.columns if df[feature].dtype != 'O']
categorical_feautres = [feature for feature in df.columns if df[feature].dtype == 'O']
```

```python
print(len(numerical_features),"Numerical Features: ",numerical_features)
print(len(categorical_feautres),"Categorical Features: ", categorical_feautres)
```

```python
for col in categorical_feautres:
    print(df[col].value_counts(normalize=True)*100)
    print("--"*100)
```

```python
for col in numerical_features:
    print(df[col].value_counts(normalize=True)*100)
    print("--"*100)
```

```python
plt.figure(figsize=(10,10))
for i in range(0,len(numerical_features)):
  plt.subplot(5,3,i+1)
  sns.kdeplot(x=df[numerical_features[i]],shade = True)
  plt.tight_layout()
```

```python
plt.figure(figsize=(20,15))
category = ["Type","Content Rating"]
for i in range(0,len(category)):
  plt.figure(figsize=(10,10))
  plt.subplot(5,3,i+1)
  sns.countplot(x=df[category[i]])
  plt.xticks(rotation=45)
  plt.tight_layout()
```
